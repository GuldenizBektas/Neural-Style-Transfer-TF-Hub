```python
import tensorflow_hub as hub
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
#Â preprocess image and load

def load_image(img_path):
  img = tf.io.read_file(img_path)
  img = tf.image.decode_image(img, channels = 3)
  img = tf.image.convert_image_dtype(img, tf.float32)
  img = img[tf.newaxis, :]
  return img
```

```python
content_image = load_image("/content/drive/MyDrive/Colab/Images/istanbul.jpg")
style_image = load_image("/content/drive/MyDrive/Colab/Images/vangogh.jpg")
```

```python
content_image.shape
```

```python
plt.imshow(np.squeeze(style_image))
plt.show()
```

```python
stylize_image = model(tf.constant(content_image), tf.constant(style_image))[0]
```

```python
plt.imshow(np.squeeze(stylize_image))
plt.show()
```

```python
cv2.imwrite("/content/drive/MyDrive/Colab/Images/generated_image.jpg", cv2.cvtColor(np.squeeze(stylize_image)*255, cv2.COLOR_BGR2RGB))
```

```python
def generate(content, style, number):
  content_image = load_image(content)
  style_image = load_image(style)

  stylized = model(tf.constant(content_image), tf.constant(style_image))[0]

  plt.imshow(np.squeeze(stylized))
  plt.show()

  cv2.imwrite("/content/drive/MyDrive/Colab/Images/generated_image{}.jpg".format(number), cv2.cvtColor(np.squeeze(stylized)*255, cv2.COLOR_BGR2RGB))

```
